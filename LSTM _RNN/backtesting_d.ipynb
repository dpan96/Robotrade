```python
buy_trades = []
sell_trades = []

for i in range(0, len(aud_d)):
    if aud_d['Prediction'][i] == 1:
        buy_trades.append((i+1, aud_d['datetime'][i+1], aud_d['open'][i+1]))
        
for i in range(0, len(aud_d)):
    if aud_d['Prediction'][i] == 0:
        sell_trades.append((i+1, aud_d['datetime'][i+1], aud_d['open'][i+1]))
    
print(buy_trades[0:10])
print(sell_trades[0:10])
```

```python
"""Set stop-loss and take-profit parameters, calculate PnL on all trades and return index of closing trade.
    Can add the timestamp in at a later stage"""
    
sl = 0.0050
tp = 0.0050
buy_trades_pnl = []
sell_trades_pnl = []

for buy in buy_trades:
    for i in range(buy[0], len(aud_d)):
        if aud_d['high'][i] >= buy[2] + tp:
            buy_trades_pnl.append((buy[0],buy[1],'buy ',buy[2],aud_d['datetime'][i],aud_d['high'][i],tp,'t/p hit',i))
            break
        elif aud_d['low'][i] <= buy[2] - sl:
            buy_trades_pnl.append((buy[0],buy[1],'buy ',buy[2],aud_d['datetime'][i],aud_d['low'][i],-sl,'s/l hit', i))
            break
        elif aud_d['Prediction'][i] == 0:
            buy_trades_pnl.append((buy[0],buy[1],'buy ',buy[2],aud_d['datetime'][i],aud_d['close'][i],
                                   (aud_d['close'][i]-buy[2]),'prediction change', i))
            break

for sell in sell_trades:
    for i in range(sell[0], len(aud_d)):
        if aud_d['low'][i] <= sell[2] - tp:
            sell_trades_pnl.append((sell[0],sell[1],'sell',sell[2],aud_d['datetime'][i],aud_d['low'][i],tp,'t/p hit',i))
            break
        elif aud_d['high'][i] >= sell[2] + sl:
            sell_trades_pnl.append((sell[0],sell[1],'sell',sell[2],aud_d['datetime'][i],aud_d['high'][i],-sl,'s/l hit',i))
            break
        elif aud_d['Prediction'][i] == 1:
            sell_trades_pnl.append((sell[0],sell[1],'sell',sell[2],aud_d['datetime'][i],aud_d['close'][i],
                                    (sell [2]-aud_d['close'][i]),'prediction change',i))
            break
```

```python

print(buy_trades_pnl[0:2])
print(buy_trades_pnl[0][8])
```

```python
good_buy_trades = [buy_trades_pnl[0]]
x = good_buy_trades[0][8]
for trade in buy_trades_pnl:
    if trade[0] > x:
        good_buy_trades.append(trade)
        x = trade[8]

good_sell_trades = [sell_trades_pnl[0]]
x = good_sell_trades[0][8]
for trade in sell_trades_pnl:
    if trade[0] > x:
        good_sell_trades.append(trade)
        x = trade[8]
```

```python
Buy_PnL = sum(x[6] for x in good_buy_trades)
Sell_PnL = sum(x[6] for x in good_sell_trades) 
trades = sorted((good_buy_trades + good_sell_trades), key=lambda x: x[0])

print('Total PnL is:', "{0:.4f}".format(Buy_PnL + Sell_PnL))
```

```python
print('entry date\t', '\t\tb/s\t', 'entry price\t', '\texit date\t', '\texit price', '  PnL    ', 'comment','\n')
for trade in trades:
    print(trade[1],' ',trade[2],'   ',"{:.4f}".format(trade[3]),'     ',trade[4],'  ',"{:.4f}".format(trade[5]),'  ',"{:.4f}".format(trade[6]),'  ',trade[7])
```

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return(TP, FP, TN, FN)
```
